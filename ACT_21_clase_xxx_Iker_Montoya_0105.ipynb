```python
# Definición de la Clase Pescador
class Pescador:
    # Constructor: Define los 3 atributos del pescador
    def __init__(self, nombre, embarcacion, experiencia):
        self.nombre = nombre               # Atributo 1
        self.embarcacion = embarcacion     # Atributo 2
        self.experiencia = experiencia     # Atributo 3

    # Función 1: Preparar la carnada
    def preparar_carnada(self):
        print(f"{self.nombre} está preparando la carnada para salir a altamar.")

    # Función 2: Lanzar la red
    def lanzar_red(self):
        print(f"{self.nombre} ha lanzado la red desde su embarcación '{self.embarcacion}'.")

    # Función 3: Recolectar la pesca
    def recolectar_pesca(self):
        print(f"{self.nombre} está subiendo los peces a bordo. ¡Hubo buena pesca!")

# --- Creación de Objetos (Instancias) ---

# Creamos el pescador1
pescador1 = Pescador("Capitán Juan", "La Niña del Mar", 15)

# Creamos el pescador2
pescador2 = Pescador("Pedro Marín", "Rayo Veloz", 5)

# --- Ejecución de acciones ---

print(f"--- Ficha Técnica: {pescador1.nombre} ---")
print(f"Bote: {pescador1.embarcacion} | Experiencia: {pescador1.experiencia} años")
pescador1.preparar_carnada()
pescador1.lanzar_red()

print("-" * 30)

print(f"--- Ficha Técnica: {pescador2.nombre} ---")
print(f"Bote: {pescador2.embarcacion} | Experiencia: {pescador2.experiencia} años")
pescador2.preparar_carnada()
pescador2.recolectar_pesca()

print("programa realizado por iker montoya 0105")
```
